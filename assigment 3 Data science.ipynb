```python

df['Order.Date'] = pd.to_datetime(df['Order.Date'])

df['Total Profit'] = df['Sales']

df['Order Month'] = df['Order.Date'].dt.month
df['Order Year'] = df['Order.Date'].dt.year

df['Profit Margin'] = df['Profit'] / df['Sales']

order_frequency = df.groupby('Customer.ID').size().reset_index(name='Order Frequency')
df = df.merge(order_frequency, on='Customer.ID', how='left')

sales_per_category = df.groupby('Category')['Sales'].transform('sum')
df['Sales per Category'] = sales_per_category

df.head()
```

```python
# COSTERMERE AVERAGE ORDER VALUE
customer_order_values = df.groupby('Customer.ID')['Sales'].agg(['sum', 'count'])

customer_order_values['Avg_Order_Value'] = customer_order_values['sum'] / customer_order_values['count']

df = df.merge(customer_order_values[['Avg_Order_Value']], on='Customer.ID', how='left')

df.head()
```

```python
# SHIPPING DATA
df['Order.Date'] = pd.to_datetime(df['Order.Date'])

df['Ship.Date'] = pd.to_datetime(df['Ship.Date'])

df['Shipping_Duration'] = (df['Ship.Date'] - df['Order.Date']).dt.days

df.head()
```
